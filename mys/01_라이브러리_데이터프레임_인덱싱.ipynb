```python
# git clone https://github.com/ranaroussi/yfinance.git
import yfinance as yf
```

```python
# 삼성전자 확인 / 국내는 http://data.krx.co.kr/contents/MDC/MAIN/main/index.cmd
samsung = yf.Ticker("005930.KS")
print(samsung.info) # 정보 확인
df1 = samsung.history(period="max")
print(df1) #period="max" 인자로 전체 기간 받아옴.
```

```python
# 테슬라 S&P 나스닥100 / 해외는 야후파이낸스
stocks = yf.Tickers(['TSLA', 'SPY', 'QQQ'])
# print(stocks.info) Tickers 는 info 함수 없음.
df2 = stocks.history(start='2020-01-01', end='2024-06-26') #기간 지정
print(df2) # Date가 인덱스
```

```python
# Ticker 종결가 검색
df1['Close']
```

```python
# Tickers 종결가 검색
df2['Close'][['TSLA','SPY']]
```

```python
# 데이터를 다룰 땐 꼭 copy 함수로 별도의 프레임을 만들어 관리하자.
df2_close = df2['Close'][['TSLA','SPY','QQQ']].copy()
df2_close
# copy를 안하면 메모리를 공유하기 때문에 원본도 수정될 위험이 있다.
```

```python
# 인덱싱 처음에서 5개
df2_close[:5]
```

```python
# 인덱싱 처음에서 5개
df2_close.head(10) #인자 안주면 5개 나옴
```

```python
# 인덱싱 마지막에서 5개
df2_close[-5:]
```

```python
# 인덱싱 마지막에서 5개
df2_close.tail(7)
```

```python
# 행 열 동시 인덱싱
df2_close[-3:][['SPY','TSLA']]
```

```python
# 인덱스이름+열이름 loc() 함수
print(df2_close.loc['2024-06-20':,:'TSLA'])

print(df2_close.loc['2024-06-20':,:'QQQ'])
```

```python
# 인덱스번호+열번호 iloc()
print(df2_close.iloc[-7:,:2])
```
